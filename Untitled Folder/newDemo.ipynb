```python
import pandas as pd
df =pd.read_csv("Book1.csv")
df.set_index("day",inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    "temperature":0,
    "windspeed":0,
    "event":"no event"
})
new_df
```

```python
new_df=df.fillna(method="ffill")
new_df
```

```python
new_df=df.interpolate(method="linear")
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
#thresh require that many non_NA value and keep that rows and remove others
new_df
```

#### Inserting Missing Dates

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx=pd.DatetimeIndex(dt)
df=new_df.reindex(idx)
df
```
