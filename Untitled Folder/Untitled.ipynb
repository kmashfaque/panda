```python
import pandas as pd
```

```python
technologies = {
    'Courses':["Spark","PySpark","Hadoop","Python","PySpark","Spark"],
    'Fee' :[20000,25000,26000,22000,24000,35000],
    'Duration':['30day','40days','35days','40days','60days','60days'],
    'Discount':[1000,2300,1200,2500,2000,2000]
              }
```

```python
df=pd.DataFrame(technologies)
df
```

```python
import numpy as np
def new_agg_func(df):
    return np.mean(df)
    
```

```python
val=df.groupby(['Courses','Fee'])
new_df=val.agg([new_agg_func,"min","max"])
new_df
```

```python
data = pd.DataFrame({"power_level": [12000, 16000, 4000, 1500, 3000, 
                                     2000, 1600, 2000, 300],
                     "uniform color": ["orange", "blue", "black", "orange",
                                       "purple", "green", "orange", "orange","orange"],
                     "species": ["saiyan","saiyan","saiyan","half saiyan",
                                 "namak","human","human","human","human"]}, 
                     index = ["Goku","Vegeta", "Nappa","Gohan",
                                   "Piccolo","Tien","Yamcha", "Krillin","Roshi"])
data
```

```python
def high_low_val(x,h,l):
    if x>=h:
        return "high"
    elif x>=l:
        return "medium"
    return "low"
    
```

```python
data["High_Low"]=data["power_level"].apply(high_low_val,args=[12000,2000])
data
```
