```python
def non_nan_mean(x):
    if x.empty:
        return None
    else: 
        x = x.dropna()
        return np.mean(x)
def standard_deviation(x):
    if x.empty:
        return None 
    else: 
        x = x.dropna()
        return np.nanstd(x)
    
def mean_lower_rating(x):
    upper = []
    for val in x:
        if val > 5:
            continue 
        else: 
            upper.append(val)
    return np.mean(upper)
```

```python
print(non_nan_mean(ser["score"]))
print(standard_deviation(ser["score"]))
print(mean_lower_rating(ser["score"]))
```

#### pandas concate 

```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df =pd.concat([india_weather,us_weather])
df
```

```python
df =pd.concat([india_weather,us_weather],ignore_index=True)
df
```

### concatenation and Keys

```python
df =pd.concat([india_weather,us_weather],keys=["india","usa"])
df
```

```python
df.loc["usa"]
```

#### concatenation using index

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])


windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

#### Concatenatiion dataframe with series

```python
s = pd.Series(["Humid","Dry","Rain"], name="event")

df=pd.concat([temperature_df,s],axis=1)
df
```
