```python
import pandas as pd

df=pd.read_excel("survey.xls")
df
```

```python
pd.crosstab(df.Nationality,df.Handedness).plot(kind="bar")
df.Age.mean()
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab(df.Nationality,df.Handedness,margins=True)
```

```python
pd.crosstab(df.Sex,[df.Nationality,df.Handedness],margins=True)
```

```python
pd.crosstab([df.Nationality, df.Sex], [df.Handedness], margins=True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, normalize=0)
```

```python
import numpy as np

pd.crosstab( df.Sex, df.Handedness, values=df.Age,aggfunc=np.average)
```
